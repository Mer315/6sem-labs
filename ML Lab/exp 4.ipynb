**Implement and demonstrate FIND-S alogrithm for finding the most specific hypothesis for a given dataset**

```python
data = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Rainy', 'Cold', 'High', 'Strong', 'Warm', 'Change', 'No'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change', 'Yes']
]

# Initialize hypothesis with the first positive example
hypothesis = ['0'] * (len(data[0]) - 1)

for instance in data:
    if instance[-1] == 'Yes':   # Consider only positive examples
        if hypothesis == ['0'] * (len(data[0]) - 1):
            hypothesis = instance[:-1]
        else:
            for i in range(len(hypothesis)):
                if hypothesis[i] != instance[i]:
                    hypothesis[i] = '?'
```

```python
print("Most specific hypothesis:/n")
print(hypothesis)
```
