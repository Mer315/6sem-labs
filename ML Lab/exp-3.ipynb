```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
def load_dataset():
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df['target'] = iris.target
    return df
```

**3a. Pandas library function for data analysis & manipulation**

```python
def pandas_operations(df):
    print("Pandas Data Analysis & Manipulation")
    print("\nFirst 5 rows of dataset:")
    print(df.head())

    print("\nDataset Information:")
    print(df.info())

    print("\nStatistical Summary:")
    print(df.describe())

    print("\nUnique Target Values:")
    print(df['target'].unique())

    print("\nMean of each feature:")
    print(df.mean())
```

**3b. Matplotlib library functions for data visualization**

```python
def matplotlib_visualization(df):
    print("\n--- Matplotlib Data Visualization ---")

    # Line Plot
    plt.figure()
    plt.plot(df['sepal length (cm)'], label='Sepal Length')
    plt.title("Line Plot of Sepal Length")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.legend()
    plt.show()

    # Scatter Plot
    plt.figure()
    plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['target'])
    plt.title("Scatter Plot: Sepal Length vs Sepal Width")
    plt.xlabel("Sepal Length")
    plt.ylabel("Sepal Width")
    plt.show()

    # Histogram
    plt.figure()
    plt.hist(df['petal length (cm)'], bins=10)
    plt.title("Histogram of Petal Length")
    plt.xlabel("Petal Length")
    plt.ylabel("Frequency")
    plt.show()
```

```python
def main():
    df = load_dataset()
    pandas_operations(df)
    matplotlib_visualization(df)
```

```python
main()
```
