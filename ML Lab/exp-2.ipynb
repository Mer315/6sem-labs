**exp 2- addition, subtraction, multiplication, division, exponentiation and modulus using Statistics, Math, NumPy, and SciPy**

```python
import math
import statistics
import numpy as np
from scipy import stats
```

```python
def get_arrays():
    n = int(input("Enter number of elements: "))
    print("Enter elements for first array:")
    arr1 = np.array([int(input()) for _ in range(n)])
    print("Enter elements for second array:")
    arr2 = np.array([int(input()) for _ in range(n)])
    return arr1, arr2
```

```python
def math_operations(arr1, arr2):
    n = len(arr1)
    add = [arr1[i] + arr2[i] for i in range(n)]
    sub = [arr1[i] - arr2[i] for i in range(n)]
    mul = [arr1[i] * arr2[i] for i in range(n)]
    div = [arr1[i] / arr2[i] for i in range(n)]
    exp = [math.pow(arr1[i], arr2[i]) for i in range(n)]
    mod = [arr1[i] % arr2[i] for i in range(n)]

    print("\nUsing Math Library:")
    print("Addition:", add)
    print("Subtraction:", sub)
    print("Multiplication:", mul)
    print("Division:", div)
    print("Exponentiation:", exp)
    print("Modulus:", mod)
```

```python
def statistics_operations(arr1, arr2):
    print("\nUsing Statistics Library:")
    print("Mean of First Array:", statistics.mean(arr1))
    print("Mean of Second Array:", statistics.mean(arr2))
    print("Sum of First Array:", sum(arr1))
    print("Sum of Second Array:", sum(arr2))
```

```python
def numpy_operations(arr1, arr2):
    print("\nUsing NumPy:")
    print("Addition:", np.add(arr1, arr2))
    print("Subtraction:", np.subtract(arr1, arr2))
    print("Multiplication:", np.multiply(arr1, arr2))
    print("Division:", np.divide(arr1, arr2))
    print("Exponentiation:", np.power(arr1, arr2))
    print("Modulus:", np.mod(arr1, arr2))
```

```python
def scipy_operations(arr1, arr2):
    print("\nUsing SciPy:")
    print("Addition:", arr1 + arr2)
    print("Subtraction:", arr1 - arr2)
    print("Multiplication:", arr1 * arr2)
    print("Division:", arr1 / arr2)
    print("Exponentiation:", arr1 ** arr2)
    print("Modulus:", arr1 % arr2)
```

```python
def main():
    arr1, arr2 = get_arrays()
    print("\nFirst Array:", arr1)
    print("Second Array:", arr2)

    math_operations(arr1, arr2)
    statistics_operations(arr1, arr2)
    numpy_operations(arr1, arr2)
    scipy_operations(arr1, arr2)
```

```python
main()
```
