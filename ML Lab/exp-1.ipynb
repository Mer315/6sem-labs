**Write a python program to compute Central Tendency Measures: Mean, Median, Mode**

```python
#without standard libs
def mean(numbers):
    mean= sum(numbers)/len(numbers)
    return mean
```

```python
def med(numbers):
    numbers.sort()
    n=len(numbers)
    if n%2==0:
        return (numbers[n//2 -1] + numbers[n//2])/2
    else:
        return numbers[n//2]
    
```

```python
def mod(numbers):
    mode = max(numbers, key=numbers.count)
    return mode
```

```python
print("Mean:", mean(numbers),"\nMedian:", med(numbers),"\nMode:", mod(numbers))
```

```python
#with libraries
import statistics

mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

**Measure of Dispersion: Variance , Standard Deviation , Interquartile range.**

```python
import math
#without standard libs
```

```python
def variance(numbers):
    return sum((x-mean(numbers))**2 for x in numbers)/ len(numbers)
    
```

```python
def stdev(numbers):
    return math.sqrt(variance(numbers))
```

```python
def iqr(numbers):
    numbers.sort()
    n=len(numbers)
    q1=numbers[n//4]
    q3=numbers[3*n //4]
    return q3-q1
```

```python
print("Variance:", variance(numbers), "\nStandard Deviation:", stdev(numbers),"\nInterQuartile Range", iqr(numbers))
```
