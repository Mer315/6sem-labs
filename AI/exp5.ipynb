# Tic Tac Toe - Two Player Game (Python)

This is a simple **2-player Tic Tac Toe** game using a 1D list as the board.

**Rules:**
- Player **X** and Player **O** take turns.
- Choose a position from **1 to 9**.
- First player to get **3 in a row** (row/column/diagonal) wins.
- If the board fills up and nobody wins â†’ **Draw**.


```python
b=[" "]*9

def show():
    print(f"\n {b[0]}|{b[1]}|{b[2]}\n -+-+-\n {b[3]}|{b[4]}|{b[5]}\n -+-+-\n {b[6]}|{b[7]}|{b[8]}\n")

def win(p):
    w=[(0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6)]
    return any(b[i]==b[j]==b[k]==p for i,j,k in w)

def play():
    p="X"; show()
    while True:
        m=int(input(f"{p} (1-9): "))-1
        if m<0 or m>8 or b[m]!=" ":
            print("Invalid move"); continue
        b[m]=p; show()
        if win(p): print(p,"wins!"); break
        if " " not in b: print("Draw!"); break
        p="O" if p=="X" else "X"

play()
```
