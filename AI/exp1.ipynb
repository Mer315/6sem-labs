# Linear Regression (from scratch)

Goal: fit a line **y = mÂ·x + c** using Gradient Descent.


```python
import numpy as np

x = np.array([1,2,3,4,5], dtype=float)
y = np.array([2,4,5,4,5], dtype=float)

m, c = 0.0, 0.0
lr = 0.01
n = len(x)

for _ in range(2000):
    yhat = m*x + c
    dm = (-2/n) * np.sum(x * (y - yhat))
    dc = (-2/n) * np.sum(y - yhat)
    m -= lr * dm
    c -= lr * dc

print("m =", round(m, 4), "c =", round(c, 4))
print("prediction for x=6:", round(m*6 + c, 4))
```
