# 8-Puzzle Solver using BFS (Breadth First Search)

This program solves the **8-puzzle problem** using **Breadth First Search (BFS)**.

### ðŸ”¹ Representation
- The board is stored as a **tuple of 9 elements** (3Ã—3 grid)
- `0` represents the **blank space**
- Goal state:


```python
from collections import deque

goal = (1, 2, 3,
        4, 5, 6,
        7, 8, 0)

moves = {"U": -3, "D": 3, "L": -1, "R": 1}

def is_valid_move(z, m):
    if m == "L" and z % 3 == 0: return False
    if m == "R" and z % 3 == 2: return False
    if m == "U" and z < 3: return False
    if m == "D" and z > 5: return False
    return True

def get_neighbors(s):
    z = s.index(0)
    ns = []
    for m in moves:
        if is_valid_move(z, m):
            np = z + moves[m]
            t = list(s)
            t[z], t[np] = t[np], t[z]
            ns.append(tuple(t))
    return ns

def bfs(start):
    q = deque([(start, [])])
    vis = {start}

    while q:
        s, path = q.popleft()
        if s == goal: return path

        for nxt in get_neighbors(s):
            if nxt not in vis:
                vis.add(nxt)
                q.append((nxt, path + [nxt]))
    return None

def print_board(s):
    for i in range(0, 9, 3):
        print(s[i], s[i+1], s[i+2])
    print()

start = (1, 2, 3,
         4, 0, 6,
         7, 5, 8)

sol = bfs(start)

if sol:
    print("Solution found in", len(sol), "moves\n")
    print("Initial State:")
    print_board(start)
    for step in sol:
        print_board(step)
else:
    print("No solution exists.")
```
